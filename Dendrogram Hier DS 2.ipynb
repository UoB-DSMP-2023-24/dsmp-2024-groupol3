# Dendograms

## Plot by Average Spend

```
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Execute SQL query in Spark and convert result to Pandas DataFrame
df = spark.sql("SELECT CAST(external_account_desc AS STRING), ROUND(AVG(amount), 2) as average_spend FROM simulated_transaction_2024_csv WHERE amount LIKE '%-%' GROUP BY external_account_desc").toPandas()

# Hierarchical clustering
X = df[['average_spend']]  # Extracting the numerical column for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Normalizing the data
mergings = linkage(X_scaled, method='complete')

# Dendrogram
plt.figure(figsize=(10, 7))

dendrogram(
    mergings,
    labels=df['external_account_desc'].tolist(),  # Convert the column to a list
    leaf_rotation=90,
    leaf_font_size=6
)
plt.title('Dendrogram of Average Spend by Type of Purchase')
plt.xlabel('Type of Purchase')
plt.ylabel('Distance')
plt.show()
```

## Plot by Avg Spend & Common Time

```
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Execute SQL query in Spark and convert result to Pandas DataFrame
df = spark.sql("""
    SELECT 
        CAST(external_account_desc AS STRING) AS external_account_desc, 
        ROUND(AVG(amount), 2) AS average_spend, 
        MODE() WITHIN GROUP (ORDER BY time_stamp) AS most_common_time
    FROM 
        simulated_transaction_2024_csv 
    WHERE 
        amount LIKE '%-%' 
        AND external_account_desc IS NOT NULL
    GROUP BY 
        external_account_desc
""").toPandas()

# Convert most common time to minutes since midnight
df['most_common_minutes'] = df['most_common_time'].apply(lambda x: int(x[:2]) * 60 + int(x[3:]))

# Extract numerical columns for clustering
X = df[['average_spend', 'most_common_minutes']]

# Normalizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Hierarchical clustering
mergings = linkage(X_scaled, method='complete')

# Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(
    mergings,
    labels=df['external_account_desc'].tolist(),
    leaf_rotation=90,
    leaf_font_size=7
    )
plt.title('Dendrogram of Average Spend and Most Common Time by Type of Purchase')
plt.xlabel('Type of Purchase')
plt.ylabel('Distance')
plt.show()

```
